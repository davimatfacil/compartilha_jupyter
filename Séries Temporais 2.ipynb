##  Estacionariedade em séries temporais

### Séries Temporais
Uma série temporal é um conjunto sequencial de pontos de dados, medido tipicamente em tempos sucessivos.

É matematicamente definido como um conjunto de vetores x(t), t = 0,1,2, ... onde t representa o tempo decorrido.

A variável x(t) é tratada como uma variável aleatória.

As medições realizadas durante um evento em uma série temporal são organizadas em uma ordem cronológica adequada.

Uma série temporal contendo registros de uma única variável é denominada como univariada e mais de uma variável como multivariada.

### Estacionaridade

- Uma série temporal é considerada estacionária se sua média e variância permanecerem constantes ao longo do tempo.Uma série desse tipo se "enrola" ao redor de uma média constante, refletindo alguma forma de equilíbrio estável. Na prática, a maioria das séries que encontramos apresentam algum tipo de não estacionariedade, por exemplo, tendência.

- Muitos modelos de séries temporais trabalham com o pressuposto de que a série temporal é estacionária. 

- Intuitivamente, podemos dizer que, se uma série temporal tem um comportamento específico ao longo do tempo, há uma probabilidade muito alta de que ela siga o mesmo no futuro.

- Além disso, as teorias relacionadas às séries estacionárias são mais consolidadas e do que outras teorias. Os modelo da classe ARIMA por exemplo utilizam esse conceito.

A estacionariedade é um importante conceito na modelagem de séries temporais e é caracterizada por uma variável que se comporta de forma aleatória ao longo do tempo ao redor de uma média constante.

Basicamente, séries temporais que possuem tendência e/ou sazonalidade não são estacionárias e é necessário o uso de técnicas adequadas a tal situação. Nesse caso modelos de alisamento expoenencial podem ser um alternativa.

### Verificando a Estacionariedade

Podemos verificar a estacionariedade de uma série temporal de algumas formas:

- **Plot das Rolling Statistics (Estatísticas Móveis)**: Podemos criar um plot da média móvel e variância móvel (ou desvio padrão) e ver se variam com o tempo. A média / variância móvel é, para qualquer instante 't', a média / variância do último ano, ou seja, os últimos 12 meses.
- **Autocorrelação com Gráficos ACF e PACF**: Se a série temporal for estacionária, os gráficos ACF / PACF (gráficos de autocorrelação da série) mostrarão um padrão específico o qual discutiremos mais adiante. 
- **Teste Dickey-Fuller**: Este é um dos testes estatísticos para verificar a estacionariedade. Aqui, a hipótese nula é que a série temporal não é estacionária. Os resultados do teste incluem uma estatística de teste e alguns valores críticos para os níveis de confiança das diferenças. Se a 'Estatística de teste' for menor que o 'valor crítico', podemos rejeitar a hipótese nula e dizer que a série é estacionária. 

Vamos analisar aqui somente  a forma gráfica de analisar estacionariedade:

### Plot das Rolling Statistics (Estatísticas Móveis)

```python
# Transformamos en array os dados
x=dados['meses'].values
y=dados['passageiros'].values
```

```python
# Determinando estatísticas móveis
rolmean = dados.rolling(window = 12).mean()
rolstd = dados.rolling(window = 12).std()
```

```python
# Plot das estatísticas móveis
orig = plt.plot(y, color = 'blue', label = 'Original')
mean = plt.plot(rolmean, color = 'red', label = 'Média Móvel')
std = plt.plot(rolstd, color = 'black', label = 'Desvio Padrão')
plt.legend(loc = 'best')
```

Pelo gráfico fica claro que a média móvel e o desvio padrão(móvel) não são constantes em relação ao tempo o que caracteriza a não estacionariedade
