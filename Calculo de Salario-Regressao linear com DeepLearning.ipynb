Neste exemplo simples, foi construido uma rede neural que preveja o salário de um funcionário com base em seus anos de experiência.

Observe que esse exemplo é apenas para fins didáticos de como usar uma rede neural para esse fim. Até porque sabemos que existem muitos outros fatores reelvantes para que se possa prever determinado salário.

Então, imagine que estimar o salário de um  funcionário fosse extremamente simples. Salário base de 50k para alguém sem experiência e cada ano adicional de experiência lhe dá um adicional de 10k.

Como você criaria uma rede neural que aprendesse esse relacionamento para prever que um funcionário com 8 anos de experiência deveria ganhar?
 Matematicamente sabemos que o valor seria 130k. Será que com uma rede neural simples podemos chegar proximo a esse valor?

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
#salary_model
def salary_model(y_new):  #cria função
    xs = [0,1,2,3,4,5,6] # numero de anos
    ys = [50,60,70,80,90,100,110] #salario em k
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])  #rede
    model.compile(optimizer="sgd", loss='mean_squared_error')  #otimizador e erro
    model.fit(xs, ys, epochs=500) #treino da rede
    return model.predict(y_new)[0][0]  #previsao
```

```python
prediction = salary_model([8.0])
```

```python
print(f'O valor previsto pela rede é {prediction:.5} k' ,)
```

Observe que o valor previsto foi 132k e não 130!!
Isso tem várias explicações. Uma delas é que o resultado se deve ao que o modelo aprendeu com 7 pontos de dados! provavelmente se forncermos mais dados a acurácia será maior!
