Vamos mostrar como fazer o gráfico de dispersão no Python usando Matplotlib e o Seaborn, dois pacotes gráficos do python

```python
#Importa pacotes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importa base de dados
dados = pd.read_csv('trees.csv')
dados.head()             
```

```python
#scatter
sns.set() #usa formato do seaborn
plt.scatter(dados['Girth'], dados['Volume'], marker='*'); #coloca as duas variaveis para os dois eixos e o tipo de marcador
plt.xlabel('Girth');
plt.ylabel('Volume');
plt.title('Arvores');
```

```python
#linhas - esse é o gráfico padrão, logo basta colocar .plot
plt.plot(dados['Girth'], dados['Volume']);
```

```python
sns.regplot(dados['Girth'], dados['Volume']); #usando seaborn, por padrao o regplot exibe a linha de regressao
```

```python
sns.regplot(dados['Girth'], dados['Volume'], x_jitter = 0.3, fit_reg = False); # se quero so os dados basta colocar False
# em fit_reg
```
